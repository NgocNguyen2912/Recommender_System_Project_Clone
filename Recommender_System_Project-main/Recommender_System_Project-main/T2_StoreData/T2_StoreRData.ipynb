#### ***1. Push raw data to Local MongoDb***

```python
# Initial step: (This step only runs in the first time)
# storing_rdata = {
#     "users_id": [],
#     "movies_id": [],
#     "umrs_id": [],
#     "murs_id": []
# }

with open("T1_CollectingData/data/storing_rdata.json", 'r') as file:
    storing_rdata = json.load(file)
storing_rdata 
```

```python
client = MongoClient('mongodb://localhost:27017/')
db = client['T2_RawData']
```

##### **1.1. Push Users Infor data**

```python
# User Infor:
## Collection
collection = db['Users_Infor']

## 
users_inf = {
    "user_id": [], 
    "user_name": [],
    "member_since": []
}

## 
directory = "T1_CollectingData/data/users_inf" #Lưu ý lấy path cho chuẩn

for filename in os.listdir(directory):
    if filename.endswith('.json'):
        filepath = os.path.join(directory, filename)
        user_id = filename[:-5]

        if user_id not in storing_rdata['users_id']:
            storing_rdata['users_id'].append(user_id)
            # Open and load the JSON file
            with open(filepath, 'r') as file:
                data_json = json.load(file)
            
            for key in users_inf.keys():
                if key in data_json:
                    users_inf[key].append(data_json[key])

with open("T1_CollectingData/data/storing_rdata.json", 'w') as json_file:
        json.dump(storing_rdata, json_file)
users_inf_df = pd.DataFrame(users_inf)
dict_df = users_inf_df.to_dict('records')
result = collection.insert_many(dict_df)
print ('=> Inserted document IDs successfuly:', result.inserted_ids) 
```

##### **1.2. Push Movies Infor data**

```python
# Movie Infor
# Collection
collection = db['Movies_Infor']

## 
movies_inf = {
    "movie_id": [],
    "title": [], 
    "introduction": [], 
    "runtime": [], 
    "runtimeSeconds": [], 
    "rating": [], 
    "award": [], 
    "genre": [], 
    "releaseDate": [], 
    "releaseLocation": [], 
    "actors": [], 
    "directors": []
}

##
directory = "T1_CollectingData/data/movies_inf" #Lưu ý lấy path cho chuẩn

for filename in os.listdir(directory):
    if filename.endswith('.json'):
        filepath = os.path.join(directory, filename)
        movie_id = filename[:-5]

        if movie_id not in storing_rdata['movies_id']:
            storing_rdata['movies_id'].append(movie_id)
            # Open and load the JSON file
            with open(filepath, 'r') as file:
                data_json = json.load(file)
            
            for key in movies_inf.keys():
                if key in data_json:
                    movies_inf[key].append(data_json[key])

with open("T1_CollectingData/data/storing_rdata.json", 'w') as json_file:
        json.dump(storing_rdata, json_file)
movies_inf_df = pd.DataFrame(movies_inf)
dict_df = movies_inf_df.to_dict('records')
result = collection.insert_many(dict_df)
print ('=> Inserted document IDs successfuly:', result.inserted_ids)
```

##### **1.3. Push (1 User - n Movies) data**
